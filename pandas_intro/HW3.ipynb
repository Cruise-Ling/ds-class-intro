# ### Exercise 5. Find the mean and standard deviation of math and verbal SAT score for men-only, women-only, and non gender specific universities.

```python
#menonly
print('sat_vr_std: {}'.format(df[df['menonly']==1]['satvrmid'].std()))
print('sat_vr_mean: {}'.format(df[df['menonly']==1]['satvrmid'].mean()))
print('sat_mt_std: {}'.format(df[df['menonly']==1]['satmtmid'].std()))
print('sat_mt_mean: {}'.format(df[df['menonly']==1]['satmtmid'].mean()))
```

```python
#womenonly
print('sat_vr_std: {}'.format(df[df['womenonly']==1]['satvrmid'].std()))
print('sat_vr_mean: {}'.format(df[df['womenonly']==1]['satvrmid'].mean()))
print('sat_mt_std: {}'.format(df[df['womenonly']==1]['satmtmid'].std()))
print('sat_mt_mean: {}'.format(df[df['womenonly']==1]['satmtmid'].mean()))
```

```python
#non-specific
print('sat_vr_std: {}'.format(df[(df['womenonly']==0) & (df['menonly']==0)]['satvrmid'].std()))
print('sat_vr_mean: {}'.format(df[(df['womenonly']==0) & (df['menonly']==0)]['satvrmid'].mean()))
print('sat_mt_std: {}'.format(df[(df['womenonly']==0) & (df['menonly']==0)]['satmtmid'].std()))
print('sat_mt_mean: {}'.format(df[(df['womenonly']==0) & (df['menonly']==0)]['satmtmid'].mean()))
```
